```python
import pandas as pd
import yfinance as yf

# Download Apple data from Yahoo Finance
apple_data = yf.download('AAPL', start="2020-01-01")

apple_data.plot(y='Close', title='AAPL Stock Price', ylabel='Price (USD)', grid=True)
apple_data["returns"]= apple_data["Close"].pct_change()
```

```python
import pandas_montecarlo
mc = apple_data['returns'].montecarlo(sims=10, bust=-0.1, goal=1) #il bust è il drawdown massimo
mc.plot(title="AAPL Returns Monte CarloSPY Simulations", figsize=(10, 6))
```
