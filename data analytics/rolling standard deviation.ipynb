```python
import pandas as pd
import numpy as np
```

```python
eurgbp_data = pd.read_csv('eurgbp data.csv', delimiter=',', index_col=0, parse_dates=True)
```

### Voglio analizzare la rolling standard deviation:

```python
import matplotlib.pyplot as plt


eurgbp_data["returns"] = eurgbp_data['close'].pct_change()
rolling_std = eurgbp_data["2024-06-15":"2024-07-02"]['returns'].rolling(window=60).std()

%matplotlib inline
import matplotlib.pyplot as plt

rolling_std.plot(figsize=(20,10), label='Rolling 60 mins standard deviation', 
                 title='EURGBP 1 min returns rolling standard deviation', xlabel='Date', ylabel='rolling Standard deviation')
```
