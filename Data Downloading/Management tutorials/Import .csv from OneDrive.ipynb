### Tutorial on how to import data from OneDrive:

Extremely useful if you need to store large files and then work with them in Python!

1) First, click on embed on OneDrive via an internet browser.
2) Second, keep only the URL from the output.
3) Third, change "embed" to "download" after onedrive.live.com/embed




```python
#This is my output from embed:

#<iframe src="https://onedrive.live.com/embed?resid=4E16D53638D067A9%21337206&authkey=!APxYFBiUDvn51MI" width="98" height="120" frameborder="0" scrolling="no"></iframe>
```

```python
import pandas as pd

#This is the URL of the file
url = "https://onedrive.live.com/download?resid=4E16D53638D067A9%21337206&authkey=!APxYFBiUDvn51MI" #REMEMBER to change embed to download;

data_frame = pd.read_csv(url, parse_dates=True, index_col=0) #it's a time series data, so I'm setting the index to be the first column
data_frame.head()
```
