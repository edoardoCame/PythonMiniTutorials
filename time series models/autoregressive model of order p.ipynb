```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
```

### Some theory:

In statistics, econometrics, and signal processing, an autoregressive (AR) model is a representation of a type of random process; as such, it can be used to describe certain time-varying processes in nature, economics, behavior, etc. The autoregressive model specifies that the output variable depends linearly on its own previous values and on a stochastic term (an imperfectly predictable term).










```python
#create a time series using the formula i provide
def create_time_series(c, yt, phi, n_simulations):

    time_series = []
    for i in range(n_simulations): #the for loop will iterate 100 times and create a time series of 100 values, remember in python the range function is exclusive of the last value!
        yt = c + yt * phi + np.random.normal(0, 1) #at every iteration the value of yt is updated, so it uses the value of yt from the previous iteration
        time_series.append(yt) #the append command adds the value to the end of the list

    #-----------------------------------------------
    #VARIOUS PLOTS
    #-----------------------------------------------
    #plot the time series
    plt.plot(time_series)
    plt.title('Time Series')
    plt.xlabel('Time')
    plt.ylabel('Value')
    
    #plot the autocorrelation
    plot_acf(time_series, lags=20)

    #print the autocorrelation
    acf(time_series, nlags=20)
    
    return time_series #return the time series

ts = create_time_series(c=0, yt=1, phi=0.5, n_simulations=1000) #create the time series
```
