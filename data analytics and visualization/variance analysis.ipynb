```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
eurgbp = pd.read_csv('/workspaces/LOCALWork/CSV data/eurgbp.csv', index_col=0, parse_dates=True)
```

```python
eurgbp["hour"] = eurgbp.index.hour
eurgbp["returns"] = eurgbp["close"].pct_change()
grouped_df = eurgbp.groupby("hour").var()

plt.bar(grouped_df.index, grouped_df["returns"])
plt.legend(["Variance of returns"])
plt.xlabel("Hour")
plt.ylabel("Variance of returns")
plt.title("Variance of returns by hour")
```
