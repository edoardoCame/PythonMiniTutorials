


The likelihood function is the joint probability of all observed data points because we want to maximize the probability of observing the entire dataset given the parameter ( p ). By considering all data points, we get a more accurate estimate of the parameter.

### Why Use the Joint Probability?
More Information: Each data point provides information about the parameter ( p ). Using all data points together gives a more robust and reliable estimate.
Law of Large Numbers: As the number of observations increases, the estimate of ( p ) becomes more accurate. This is due to the law of large numbers, which states that the average of the results obtained from a large number of trials should be close to the expected value.
Consistency: The MLE method is consistent, meaning that as the sample size increases, the MLE converges to the true parameter value.
Example with One Data Point vs. Multiple Data Points
One Data Point: If you only use one data point, the estimate of ( p ) would be either 0 or 1, depending on whether the data point is 0 or 1. This is not a reliable estimate.
Multiple Data Points: Using multiple data points, you can calculate the proportion of successes (1s) to the total number of trials, which gives a more accurate estimate of ( p ).

The Methods of Moments (MoM) and Maximum Likelihood Estimation (MLE) are both techniques used for parameter estimation in statistics, but they have different approaches and properties.

### Methods of Moments (MoM)
1. **Approach**:
   - MoM involves equating sample moments (e.g., sample mean, sample variance) to theoretical moments (e.g., population mean, population variance) and solving for the parameters.
   - For example, if you have a sample mean \(\bar{x}\) and you know the theoretical mean \(\mu\) is a function of the parameter \(\theta\), you set \(\bar{x} = \mu(\theta)\) and solve for \(\theta\).

2. **Properties**:
   - Simpler to compute, especially for small samples.
   - May not always provide efficient or unbiased estimates.
   - Does not require the specification of a likelihood function.

### Maximum Likelihood Estimation (MLE)
1. **Approach**:
   - MLE involves finding the parameter values that maximize the likelihood function, which measures how likely it is to observe the given sample data given the parameters.
   - For example, if \(L(\theta)\) is the likelihood function, you find \(\hat{\theta}\) such that \(L(\hat{\theta})\) is maximized.

2. **Properties**:
   - Often provides efficient and unbiased estimates, especially for large samples (asymptotically efficient).
   - Requires the specification of a likelihood function, which can be complex for some distributions.
   - Estimates are consistent and normally distributed for large samples.

### Summary
- **MoM**: Simpler, based on matching sample and theoretical moments, may not be efficient or unbiased.
- **MLE**: More complex, based on maximizing the likelihood function, generally provides efficient and unbiased estimates for large samples.

Both methods have their own advantages and are chosen based on the specific requirements and constraints of the problem at hand.

```python
import numpy as np
from scipy.stats import norm
from scipy.optimize import minimize

# Sample data
sample_data = np.random.normal(1, 2, 1000)

# Define the likelihood function for a normal distribution
def likelihood(params):
    mu, sigma = params
    return np.prod(norm.pdf(sample_data, loc=mu, scale=sigma))

# Define the negative log-likelihood function (to be minimized)
def neg_log_likelihood(params):
    return -np.log(likelihood(params))

# Initial parameter values
initial_params = [1, 1]

# Minimize the negative log-likelihood to find the MLE estimates
result = minimize(neg_log_likelihood, initial_params)

# Extract the estimated parameters
estimated_mu, estimated_sigma = result.x

# Print the estimated parameters
print("Estimated mu:", estimated_mu)
print("Estimated sigma:", estimated_sigma)


### PROBLEM WITH DIVISION BY ZERO IN -NP.LOG(LIKELIHOOD(PARAMS))
############################################################################################################
```
